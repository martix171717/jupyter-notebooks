```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
prices = [
    (1, 2.12),
    (2, 2.56),
    (3, 3.10),
    (4, 3.16),
    (5, 3.58),
    (6, 5.12),
    (7, 5.16),
    (8, 5.20),
    (9, 4.12),
    (10, 4.10),
    (11, 3.65),
    (12, 4.25),
]
```

```python
df = pd.DataFrame(prices)
df
```

```python
df = pd.DataFrame(prices, columns=["month", "PLN_price"])
df=df.set_index('month')
df
```

```python
df["USD_price"]=df["PLN_price"]/4.00   # 1 USD = 4 PLN
df
```

```python
plt.xlabel('month')
plt.title('USD price of goods')
plt.ylim(0, 2.00)
plt.grid(True)
plt.plot(df.index, df["USD_price"], 'r--')
```
